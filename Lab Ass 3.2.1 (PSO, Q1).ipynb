```python
import numpy as np
import matplotlib.pyplot as plt
pop=np.random.rand(10,1)*(2*np.pi)
def cfun(a):
        return np.sin(a)
def Gbest(x):
    Cfun=np.apply_along_axis(cfun,1,x)
    return np.argmin(Cfun)
gbest=Gbest(pop)
pbest=np.copy(pop)
v=np.zeros((10,1))
GBests=np.zeros((201,1))
GBests[0]=gbest
for i in range(0,200):
    for j in range(0,pop.shape[0]):
        v[j]+=(2*(np.random.rand())*(pbest[j]-pop[j]))+(2*(np.random.rand())*(gbest-pop[j]))
        z=pop[j]+v[j]
        if z<=0:
            pop[j]=0
        elif z>=(2*np.pi):
            pop[j]=2*np.pi
        else:
            pop[j]=z
        if (np.sin(pop[j]))<(np.sin(pbest[j])):
            pbest[j]=pop[j]
        else:
            pass
        
    gbest=pbest[Gbest(pbest)]
    GBests[i+1]=gbest
print(GBests)
GBestsfun=np.apply_along_axis(cfun,1,GBests)
GBestsfun
```
