```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x=np.random.randint(1,100,[500,2])
def fun(a):
    return np.mean(a)
mean=np.apply_along_axis(fun,0,x)
x_u=x-mean
x_u_t=np.transpose(x_u)
shape=x.shape[0]
cov=(np.matmul(x_u_t,x_u))/shape
inv=np.linalg.inv(cov)
det=np.linalg.det(cov)
p=np.empty(500)
for i in range(0,500):
    e=np.matmul((x[i]-mean),inv)
    r=np.matmul(e,np.transpose(x[i]-mean))*-0.5
    d=(2*(math.pi)*(math.sqrt(det)))
    fun=(math.e**r)/d
    p[i]=fun
fig=plt.figure()
ax=fig.gca(projection='3d')
ax.plot_trisurf(x[:,0],x[:,1],p,linewidth=0.2)
surf=ax.plot_trisurf(x[:,0],x[:,1],p, cmap=plt.cm.viridis, linewidth=0.2)
fig.colorbar( surf, shrink=0.5, aspect=5)
plt.show();  
```
