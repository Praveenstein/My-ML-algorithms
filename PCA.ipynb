```python
import pandas as pd
import numpy as np
d=pd.read_csv('ionosphere.txt',header=None)
data=np.array(d.iloc[:,:34])
def fun(a):
    return np.mean(a)
mean=np.apply_along_axis(fun,0,data)
print(mean.shape)
data_u=data-mean
data_u_t=np.transpose(data_u)
shape=data.shape[0]
cov=(np.matmul(data_u_t,data_u))/shape
e,v=np.linalg.eig(cov)
ei=np.argsort(-e)
e=np.sort(-e)*-1
su=np.sum(e)
v=v[:,ei]
eigenvector=pd.DataFrame(v,columns=e)
pb=0
k=0
for i in range(0,e.size):
    if pb<0.9:
        pb+=e[i]/su
        k=i+1
reduced_data=np.matmul(data,v[:,0:k])
reduced_data_frame=pd.DataFrame(reduced_data)
reduced_data_frame.head()
```
